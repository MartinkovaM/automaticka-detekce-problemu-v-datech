```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Načtení datasetu
df = pd.read_csv("../data/AirQualityUCI.csv", sep=';', decimal=',')

# Odstranění prázdných sloupců
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Odstranění prázdných řádků
df.dropna(how='all', inplace=True)

df.head()
```

```python
# Základní informace o datech
print(f"Počet řádků: {df.shape[0]}, Počet sloupců: {df.shape[1]}")
```

```python
# Deskriptivní statistika
df.describe().round(2)
```

```python
# Histogramy vybraných atributů
cols = ['CO(GT)', 'NOx(GT)', 'NO2(GT)', 'C6H6(GT)', 'T']

for col in cols:
    plt.figure()
    df[col].hist(bins=50)
    plt.title(f"Distribuce: {col}")
    plt.xlabel(col)
    plt.ylabel("Počet výskytů")
    plt.grid(True)
```
